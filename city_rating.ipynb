```
from pyspark import SparkConf, SparkContext
```

```
conf = SparkConf().setAppName('Rating')
sc = SparkContext.getOrCreate(conf)
```

```
text_RDD = sc.textFile('/FileStore/tables/city_rating.txt')
```

```
split_rdd = text_RDD.map(lambda x: (x.split(',')[0],(float(x.split(',')[2]), 1)))
split_rdd.collect()
```

```
reduced_rdd = split_rdd.reduceByKey(lambda x,y: (x[0]+y[0], x[1]+ y[1]))
reduced_rdd.collect()
```

```
avg_rdd = reduced_rdd.map(lambda x: (x[0], x[1][0]/x[1][1]))
avg_rdd.collect()
```
