```
from pyspark.sql import SparkSession
from pyspark.sql.functions import count, sum, min, max, mean
```

```
spark = SparkSession.builder.appName('group filter').getOrCreate()
df = spark.read.options(header=True, inferSchema=True).csv('/FileStore/tables/StudentData.csv')
```

```
df.printSchema()
```

```
df.groupBy('course').agg(count(col="*")).show()
```

```
df.groupBy('course','gender').agg(count(col="*")).show()
```

```
df.groupBy('course','gender').agg(sum('marks')).show()
```

```
df.groupBy('age', 'course').agg(min('marks'), max('marks'), mean('marks')).sort('age', 'course').show()
```
