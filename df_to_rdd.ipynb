```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('spark df rdd').getOrCreate()
```

```
df = spark.read.options(header=True, inferSchema=True).csv('/FileStore/tables/OfficeData.csv')
```

```
df.show()
```

```
rdd = df.rdd
```

```
rdd.collect()
```
