```
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf, lit
from pyspark.sql.types import IntegerType, FloatType
```

```
spark = SparkSession.builder.appName('UDF').getOrCreate()
df = spark.read.options(header=True, inferSchema=True).csv('/FileStore/tables/OfficeData.csv')
```

```
df.show()
```

```
def increase(salary, bonus, state):
    inc = 10*salary/100 + 5*bonus/100 if state == 'NY' else 12*salary/100 + 3*bonus/100
    return inc
```

```
increament = udf(lambda x,y,z: increase(x,y,z), FloatType())
```

```
df = df.withColumn('increament', increament(df.salary, df.bonus, df.state))
```

```
df.show()
```
