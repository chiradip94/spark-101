```
from pyspark import SparkConf, SparkContext
```

```
conf = SparkConf().setAppName("Simple Map")
sc = SparkContext.getOrCreate(conf)
```

```
fileRDD = sc.textFile('/FileStore/tables/lines.txt')
```

```
mapRDD = fileRDD.flatMap(lambda x: x.split())
```

```
filteredRDD = mapRDD.filter(lambda x: (x[0] != 'H' and x[0] != 'y'))
```

```
filteredRDD.collect()
```
