```
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```
spark = SparkSession.builder.appName('DF Group').getOrCreate()
df = spark.read.options(header=True, inferSchema=True).csv('/FileStore/tables/StudentData.csv')
df.show()
```

```
df.groupBy('gender').sum('marks').show()
df.groupBy('gender').mean('marks').show()
df.groupBy('gender').avg('marks').show()
df.groupBy('gender').max('marks').show()
df.groupBy('course','gender').min('marks').show()
```

```
df.groupBy('course').agg(count("*"), sum('marks'), mean('marks'), max('marks').alias('Leader Board')).show()
```

```
df.groupBy('course','gender').agg(count("*"), sum('marks'), mean('marks'), max('marks').alias('Leader Board')).show()
```
