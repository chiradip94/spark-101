```
from pyspark import SparkConf, SparkContext
```

```
conf = SparkConf().setAppName('Character Count')
```

```
sc = SparkContext.getOrCreate(conf)
```

```
rdd1 = sc.textFile('/FileStore/tables/lines.txt')
```

```
def count_chars(x):
    lst = x.split()
    cnt_lst = []
    for w in lst:
        cnt_lst.append(len(w))
    return cnt_lst
```

```
rdd2 = rdd1.map(count_chars)
rdd2.collect()
```
