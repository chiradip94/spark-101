```
from pyspark import SparkConf, SparkContext
```

```
conf = SparkConf().setAppName("Proj 1")
sc = SparkContext.getOrCreate(conf)
file_rdd = sc.textFile('/FileStore/tables/StudentData.csv')
```

```
map = file_rdd.map(lambda x: x.split(','))
head = map.first()
```

```
datas_rdd = map.filter(lambda x: x != head)
```

```
datas_rdd.count()
```

```
datas_rdd.map(lambda x: (x[1], int(x[5]))).reduceByKey(lambda x,y: x+y).collect() #Total marks
```

```
datas_rdd.filter(lambda x: int(x[5]) >= 50).count() #passed
```

```
datas_rdd.filter(lambda x: int(x[5]) < 50).count() #failed
```

```
datas_rdd.map(lambda x: x[3]).countByValue() #student enrolled per course
```

```
datas_rdd.map(lambda x: (x[3], int(x[5]))).reduceByKey(lambda x,y: x+y).collect() #Tot Marks per sub
```

```
datas_rdd.map(lambda x: (x[3], (int(x[5]), 1))).reduceByKey(lambda x,y: (x[0]+y[0], x[1]+y[1])).map(lambda x: (x[0],x[1][0]/x[1][1])).collect() #Avg marks per sub
```

```
datas_rdd.map(lambda x: (x[3], int(x[5]))).reduceByKey(lambda x,y: x if x>y else y).collect() #Max marks per sub
```

```
datas_rdd.map(lambda x: (x[3], int(x[5]))).reduceByKey(lambda x,y: x if x<y else y).collect() #Min marks per sub
```

```
datas_rdd.map(lambda x: (x[1], (int(x[0]), 1))).reduceByKey(lambda x,y: (x[0]+y[0], x[1]+y[1])).map(lambda x: (x[0],x[1][0]/x[1][1])).collect() #Avg age of male and female student
```
