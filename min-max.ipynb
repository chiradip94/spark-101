```
from pyspark import SparkConf, SparkContext
```

```
conf = SparkConf().setAppName("Min Max Rate")
sc = SparkContext.getOrCreate(conf)
```

```
file_rdd = sc.textFile('/FileStore/tables/city_rating.txt')
```

```
city_map = file_rdd.map(lambda x: (x.split(',')[1],(x.split(',')[2])))
city_map.collect()
```

```
max_rdd = city_map.reduceByKey(lambda x,y: x if x>y else y)
max_rdd.collect()
```

```
min_rdd = city_map.reduceByKey(lambda x,y: x if x<y else y)
min_rdd.collect()
```
