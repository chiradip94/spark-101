```
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit, col
```

```
spark = SparkSession.builder.appName('DF Assignment 1').getOrCreate()
df = spark.read.options(header=True, inferSchema=True).csv('/FileStore/tables/StudentData.csv')
```

```
df.show()
```

```
df = df.withColumn(colName='total marks', col=lit(120))
df.show()
```

```
df = df.withColumn(colName='average', col=(df.marks/col('total marks'))*100)
df.show()
```

```
df_oop_pass = df.filter((col('course') == 'OOP') & (col('average') > 80))
df_oop_pass.show()
```

```
df_cloud_pass = df.filter((col('course').startswith('Cl') ) & (col('average') > 60))
df_cloud_pass.show()
```

```
df_cloud_pass.union(df_oop_pass).select('name', 'marks').show()
```
