```
from pyspark import SparkConf, SparkContext
```

```
conf = SparkConf().setAppName("Word count")
sc = SparkContext.getOrCreate(conf)
```

```
file_RDD = sc.textFile('/FileStore/tables/lines.txt')
```

```
wordsRDD_flat = file_RDD.flatMap(lambda x: (x.split()))
wordsRDD = wordsRDD_flat.map(lambda x: (x,1))
```

```
#wordsRDD.collect()
wordsRDD.reduceByKey(lambda x,y: x+y).collect()
```
